```python
# Exemplo de uso - AINDA NÃO ESTÁ PRONTO a arvore com o print ainda tá ruim
texto = "AATGCC"
frequencias = {"A": 40, "T": 35, "G": 10, "C": 15}

huffman = Huffman()
comprimido_huffman, codificacao = huffman.comprimir(texto, frequencias)
descomprimido_huffman = huffman.descomprimir(comprimido_huffman, codificacao)

print(f"Texto original: {texto}")
print(f"Texto comprimido (Huffman): {comprimido_huffman}")
print(f"Texto descomprimido (Huffman): {descomprimido_huffman}")
```
