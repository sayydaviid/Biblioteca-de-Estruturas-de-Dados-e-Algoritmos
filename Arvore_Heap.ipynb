```python
from modulo_arvores import Heap
from objects import Logger, utils
import random
Logger.get_instance().activate()
```

```python
# Exemplo 1: Construindo o Heap Máximo com a lista de números fornecidos
heap_max = Heap(tipo="max")
numeros = [11, 30, 8, 10, 24, 22, 5, 12, 1, 20, 13, 23, 18, 3]

# Inserindo os números no heap
for numero in numeros:
    heap_max.inserir(numero)

# Log para informar que a exibição do heap está prestes a acontecer
Logger.get_instance().log("Heap Máximo após a inserção dos números:")

heap_max.exibir_arvore() 
```

```python
# Exemplo 2: Alterando as prioridades
heap_max.alterar_prioridade(30, 4)  # Alterando a prioridade de 30 para 4
heap_max.alterar_prioridade(1, 31)  # Alterando a prioridade de 1 para 31
heap_max.alterar_prioridade(10, 29)  # Alterando a prioridade de 10 para 29
```

```python
# Exemplo 3: Removendo 3 elementos prioritários
Logger.get_instance().log("Removendo 3 elementos prioritários:")
heap_max.remover()
heap_max.remover()
heap_max.remover()
```
